Check for Balanced Brackets in an expression
Given an expression string **exp**, write a program to examine whether the pairs and the orders of “{“, “}”, “(“, “)”, “[“, “]” are correct in the given expression.

Conditons:


Stack class from Basic tasks must be used
`is_brackets_balanced(self, expr: str)->bool` must be Stack's method
Examples:


```python
> “[(abc)]{a}{[(b)(42)]*(56)}"
True


> “[(abc])” 
False
```

````python
class Stack:
    def __init__(self):
        # Initialize an empty list to represent the stack
        self.stack = []

    def push(self, item):
        # Add an item to the top of the stack
        self.stack.append(item)

    def pop(self):
        # check if the stack is not empty before popping an item
        if not self.is_empty():
            # Remove and return the item at the top of the stack
            return self.stack.pop()
        else:
            print("Stack is empty")

    # Return the item at the top of the stack without removing it
    def peek(self):
        # check if the stack is not empty before peeking
        if not self.is_empty():
            # Return the item at the top of the stack without removing it
            return self.stack[-1]
        else:
            print("Stack is empty")

    def is_empty(self):
        # Returns `True` if there are no elements.
        return len(self.stack) == 0

    def display(self, message: str = None):
        print(
            (f"{message}: " if message else "") +
            f"{self.stack}"
        )

    # This method checks if the brackets in the given expression are balanced.
    def is_brackets_balanced(self, expr: str) -> bool:
        # A dictionary of matching closing parentheses to opening parentheses.
        pairs = {')': '(', ']': '[', '}': '{'}

        # Creating sets of opening and closing brackets for quick lookup.
        openings = set(pairs.values())
        closings = set(pairs.keys())

        for ch in expr:
            # if the character is an opening bracket, if so, push it onto the stack
            if ch in openings:
                self.push(ch)

            elif ch in closings:
                # if the stack is empty before popping an item, if it is empty, then the brackets are not balanced
                if self.is_empty():
                    return False
                
                # Take the last opening parenthesis.
                top = self.pop()

                # if it doesn't the closing parenthesis matches the last opening parenthesis, then the brackets are not balanced
                if top != pairs[ch]:
                    return False

        return self.is_empty()


if __name__ == "__main__":
    expr = input().strip()
    stack = Stack()
    print(stack.is_brackets_balanced(expr))


# === Expected Input and Output ===

# > “[(abc)]{a}{[(b)(42)]*(56)}"
# True

# > “[(abc])” 
# False
````
