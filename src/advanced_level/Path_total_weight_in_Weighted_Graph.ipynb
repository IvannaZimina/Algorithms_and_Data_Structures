#### Goal
Implement in existing class `get_path_weight()` method returning the total weight of edges on the input path  

Conditions  

You have the WeightedGraph class already implemented.  
If there is no path in graph return -1  
In addition to class method your script must have code processing the following input:  
```python
> 5 # Total num of vertices
> 2 # Num of edges to add
> 1 2 15 # 1st
> 2 3 13 # 2nd
> 1 2 3 # Path
28
```

````python
from abc import ABC, abstractmethod


class GraphInterface(ABC):
    @abstractmethod
    def __init__(self, num_vertices: int) -> None:
        super().__init__()

    @abstractmethod
    def add_edge(v1: int, v2: int) -> None:
        pass

    @abstractmethod
    def is_adjacent(v1: int, v2: int) -> bool:
        pass

    @abstractmethod
    def return_adjacent(v: int) -> tuple[int, ...]:
        pass


class WeightedGraphInterface(GraphInterface):
    @abstractmethod
    def add_edge(v1: int, v2: int, w: int) -> None:
        pass

    @abstractmethod
    def get_path_weight(*path) -> int:
        """Traverse a path and calculates total weight

        Returns:
            int: -1 if there is no such a path in graph
            total weight of path otherwise
        """
        pass


class WeightedGraph(WeightedGraphInterface):
    def __init__(self, num_vertices: int) -> None:
        super().__init__(num_vertices)
        self.data = [
            [0] * num_vertices for _ in range(num_vertices)
        ]

    def add_edge(self, v1: int, v2: int, w: int) -> None:
        try:
            self.data[v1][v2] = w
            self.data[v2][v1] = w
        except IndexError as exc:
            raise Exception(
                f"There is no {v1} or {v2} in the graph with {len(self.data)} vertices"
            ) from exc

    def is_adjacent(self, v1: int, v2: int) -> bool:
        return self.data[v1][v2] > 0

    def return_adjacent(self, v: int) -> tuple[int, ...]:
        return tuple(i for i, x in enumerate(self.data[v]) if x > 0)

    def get_path_weight(self, *path) -> int:
        if len(path) < 2:
            return 0

        total = 0
        try:
            for v1, v2 in zip(path, path[1:]):
                if not self.is_adjacent(v1, v2):
                    return -1
                total += self.data[v1][v2]
        except IndexError:
            return -1

        return total


if __name__ == "__main__":
    """> 5      # Total num of vertices
       > 2      # Num of edges to add
       > 1 2 15 # 1st
       > 2 3 13 # 2nd
       > 1 2 3  # Path
       28
    """
    # get Total num of vertices
    graph = WeightedGraph(int(input().strip()))

    # get Num of edges to add
    num_edges = int(input().strip())
    for _ in range(num_edges):
        v1, v2, w = map(int, input().split())
        graph.add_edge(v1, v2, w)

    # get a path
    path = tuple(map(int, input().split()))

    # return total weight of the path
    print(graph.get_path_weight(*path))
````
