# Implement a Circular Linked List in Python

## Objective
Implement a Circular Linked List with fundamental operations, ensuring that the last node connects back to the head.

## Requirements
Define a `Node` class with:

- A `value` attribute to store data.
- A `next` pointer linking to the next node.

Create a `CircularLinkedList` class that includes:

- `insert_at_head(value)`: Insert a new node at the beginning.
- `insert_at_tail(value)`: Insert a new node at the end.
- `delete(value)`: Remove a node with the given value.
- `display()`: Print all elements in the list, looping back to the head if necessary.

## Test the Implementation

- Insert multiple elements and verify circular linkage.
- Delete nodes and check if the circular property remains intact.
- Handle edge cases, such as an empty list or deleting the only node.

## Expected Outcome
The program should maintain a properly functioning circular linked list, allowing dynamic insertion, deletion, and traversal.
