# Implement a Circular Queue Using an Array

## Objective
Complete the implementation of a Circular Queue using a fixed-size array (Python list) without changing its length.

## Requirements
Initialize the queue with a fixed capacity.

Implement core operations:
- `enqueue(value)`: Add an element if space is available.
- `dequeue()`: Remove and return the front element.
- `peek()`: Return the front element without removing it.
- `is_empty()` / `is_full()`: Check queue status.
- `display()`: Print all queue elements in order.

Follow the given template with `# TODO:` placeholders to complete missing logic.

## Expected Outcome
A fully functional Circular Queue where elements wrap around when reaching the end of the array.

## Additional Conditions

- python `list` can be used as array for simplicity
- it is prohibited to change length of list
