# Implement a Singly Linked List in Python

## Objective
Implement a Singly Linked List with essential operations such as insertion, deletion, and traversal.

## Requirements
Define a `Node` class with:

- A `value` attribute to store data.
- A `next` pointer to link nodes.

Create a `LinkedList` class that includes:

- `insert_at_head(value)`: Insert a new node at the beginning.
- `insert_at_tail(value)`: Insert a new node at the end.
- `delete(value)`: Remove a node with the given value.
- `display()`: Print the linked list elements.

## Test the implementation

- Insert multiple elements and verify the list structure.
- Delete specific elements and check correctness.
- Ensure edge cases like deleting from an empty list or removing the only node are handled.

## Expected Output
Your program should correctly maintain the linked list structure, allowing dynamic modifications and proper traversal.
