# Implement a Deque Using an Array

## Objective
Complete the implementation of a Deque (Double-Ended Queue).

## Requirements
Initialize the deque with a fixed capacity.

Implement core operations:
- `insert_front(value)`: Add an element at the front.
- `insert_rear(value)`: Add an element at the rear.
- `delete_front()`: Remove and return the front element.
- `delete_rear()`: Remove and return the rear element.
- `peek_front()` / `peek_rear()`: View elements without removal.
- `is_empty()` / `is_full()`: Check deque status.
- `display()`: Print all elements in order.

Follow the given template with `# TODO:` placeholders to complete missing logic.

## Expected Outcome
A fully functional Deque where elements can be inserted and removed from both ends efficiently.
