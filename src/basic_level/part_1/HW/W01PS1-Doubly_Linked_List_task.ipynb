# Implement a Doubly Linked List in Python

## Objective
Implement a Doubly Linked List (DLL) that supports bidirectional traversal and key operations such as insertion, deletion, and display.

## Requirements
Define a Node class with:

- A value attribute to store data.
- A next pointer linking to the next node.
- A prev pointer linking to the previous node.

Create a DoublyLinkedList class that includes:

- insert_at_head(value): Insert a new node at the beginning.
- insert_at_tail(value): Insert a new node at the end.
- delete(value): Remove a node with the given value.
- display_forward(): Print the list from head to tail.
- display_backward(): Print the list from tail to head.

## Test the Implementation

- Insert multiple elements and verify both forward and backward traversal.
- Delete nodes and ensure the list maintains proper bidirectional links.
- Handle edge cases like an empty list or deleting the only node.

## Expected Outcome
The program should correctly maintain a doubly linked list, allowing efficient insertion, deletion, and traversal in both directions.
