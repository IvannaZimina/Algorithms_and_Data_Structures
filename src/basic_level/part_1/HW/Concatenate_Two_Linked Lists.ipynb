# Concatenate Two Linked Lists
If we append the second linked list to the end of the first linked list, it's called concatenation.

## Conditons:

- LinkedList class made on Practical Session 1 must be used
- concat(ll: LinkedList)->LinkedList must be LinkedList's method

## Example:

```python
> ll1.display()
> ll2.display()
> ll = ll1.concat(ll2)
> ll.display
2->9->6->8
10->4->7
2->9->6->8->10->4->7
```
